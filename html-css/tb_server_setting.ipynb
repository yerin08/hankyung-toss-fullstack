```python
%%writefile server.js

const express = require("express");
const path = require("path");
const app = express();

/* ============================================================
   1. ž†ēž†Ā ŪĆĆžĚľ(žĚīŽĮłžßÄ, CSS, JS) žĄ§ž†ē
   - 'static' ŪŹīŽćĒ žēąžóź žěąŽäĒ ŪĆĆžĚľŽď§žĚĄ "/static" Í≤ĹŽ°úŽ°ú ž†úÍ≥ĶŪē©ŽčąŽč§.
   - žĚīž†ú HTMLžóźžĄú žĚīŽĮłžßÄŽ•ľ Ž∂ąŽü¨žė¨ ŽēĆ <img src="/static/logo.png"> Ž°ú žā¨žö©ŪēėžĄłžöĒ.
   ============================================================ */
app.use("/static", express.static(path.join(__dirname, "static")));


/* ============================================================
   2. HTML ŽĚľžöįŪäł (templates ŪŹīŽćĒ Žāī ŪĆĆžĚľ žóįÍ≤į)
   - žā¨žö©žěźžĚė žöĒž≤≠žóź ŽĒįŽĚľ templates ŪŹīŽćĒžĚė HTMLžĚĄ ž†ĄžÜ°Ūē©ŽčąŽč§.
   ============================================================ */

const TEMPLATE_PATH = path.join(__dirname, "templates");

// [1] Home
app.get("/", (req, res) => {
  res.sendFile(path.join(TEMPLATE_PATH, "tossbank.html"));
});

// [2] About
app.get("/about", (req, res) => {
  res.sendFile(path.join(TEMPLATE_PATH, "tossbank_about.html"));
});

// [3] Products
app.get("/products", (req, res) => {
  res.sendFile(path.join(TEMPLATE_PATH, "tossbank_products.html"));
});

// [4] Loan
app.get("/loan", (req, res) => {
  res.sendFile(path.join(TEMPLATE_PATH, "tossbank_loan.html"));
});

// [5] Support
app.get("/support", (req, res) => {
  res.sendFile(path.join(TEMPLATE_PATH, "tossbank_support.html"));
});

/* ============================================================
   3. žĄúŽ≤Ą žč§ŪĖČ
   ============================================================ */
app.listen(3000, () => {
  console.log("Server is running on http://localhost:3000");
  console.log("HTML location: ./templates");
  console.log("Static location: ./static");
});
```
