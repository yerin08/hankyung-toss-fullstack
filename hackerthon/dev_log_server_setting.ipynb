```python
%%writefile server.js

const express = require('express');
const path = require('path');
const app = express();
const PORT = 3000;

// 1. Ï†ïÏ†Å ÌååÏùº ÏÑ§Ï†ï (CSS, JS, Ïù¥ÎØ∏ÏßÄ, Ìè∞Ìä∏ Îì±)
app.use('/static', express.static(path.join(__dirname, 'static')));

// 2. ÎùºÏö∞ÌåÖ ÏÑ§Ï†ï
// Î©îÏù∏ Ìôà (Home)
app.get('/', (req, res) => {
  res.sendFile(path.join(__dirname, 'templates', 'home.html'));
});

// MBTI ÌÖåÏä§Ìä∏ ÌéòÏù¥ÏßÄ
app.get('/mbti', (req, res) => {
  res.sendFile(path.join(__dirname, 'templates', 'mbti.html'));
});

// ÏïÑÌã∞ÌÅ¥ ÌéòÏù¥ÏßÄ
app.get('/article', (req, res) => {
  res.sendFile(path.join(__dirname, 'templates', 'article.html'));
});

// 3. ÏÑúÎ≤Ñ Ïã§Ìñâ
app.listen(PORT, () => {
  console.log(`================================================`);
  console.log(` Dev-Log ÏÑúÎ≤ÑÍ∞Ä Ìè¨Ìä∏ ${PORT}ÏóêÏÑú ÏûëÎèô Ï§ëÏûÖÎãàÎã§.`);
  console.log(`================================================`);
});
```
