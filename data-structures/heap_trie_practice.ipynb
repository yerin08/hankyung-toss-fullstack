# 힙 (Heap)

### 1. Heap의 정의와 목적

**정의**

Heap(힙)은 완전 이진 트리(Complete Binary Tree) 구조를 기반으로 부모 노드와 자식 노드 간의 대소 관계를 유지하는 부분 정렬 자료구조로서 전체 데이터가 정렬되어 있지는 않지만 루트 노드에는 항상 최댓값 또는 최솟값이 위치하도록 설계된 구조이며 동시에 컴퓨터 메모리 구조에서 Heap 영역은 실행 중 필요한 데이터를 동적으로 할당하고 해제하는 공간으로 사용되며 프로그램 실행 시점에 크기가 결정되지 않는 객체나 배열과 같은 참조 타입 데이터가 저장되는 영역이다.

**목적**

- **우선순위 처리:** 데이터 전체를 정렬하지 않고도 가장 중요한 값을 빠르게 추출하기 위해 사용됩니다.
- **효율성:** 삽입과 삭제를 O(log n)으로 처리하면서 최상위 값 접근을 O(1)로 유지합니다.
- **동적 데이터 처리:** 실시간으로 데이터가 추가되는 환경에서도 정렬 상태를 유지할 수 있습니다.
- **동적 메모리 관리:** Heap 메모리 영역은 실행 중 필요한 만큼 메모리를 할당하고 해제할 수 있도록 하여 객체, 배열, 함수와 같은 참조 타입 데이터를 유연하게 관리할 수 있도록 하며 이는 Stack과 달리 크기가 유동적인 데이터 구조를 처리하기 위한 핵심 메커니즘이다.

### 2. Heap의 구조와 메모리 표현

Heap은 트리 구조이지만 실제 구현은 배열로 수행되며 이는 메모리 접근 효율을 높이고 포인터 없이 구조를 유지하기 위함입니다.

- 부모 노드 인덱스: `Math.floor((i - 1) / 2)`
- 왼쪽 자식 인덱스: `2i + 1`
- 오른쪽 자식 인덱스: `2i + 2`

### 3. 최소 힙과 최대 힙

**최소 힙 (Min Heap)**

- 부모 ≤ 자식 관계를 유지하며 루트에는 항상 최소값이 위치합니다.

**최대 힙 (Max Heap)**

- 부모 ≥ 자식 관계를 유지하며 루트에는 항상 최대값이 위치합니다.

### 4. 핵심 알고리즘 (Heapify)

**Heapify Up**

삽입 시 수행되며 새로운 데이터를 마지막 위치에 삽입한 후 부모 노드와 비교하여 우선순위 조건을 만족할 때까지 위로 이동시키는 알고리즘입니다.

**Heapify Down**

삭제 시 수행되며 루트 노드를 제거한 후 마지막 노드를 루트로 이동시키고 자식 노드와 비교하여 우선순위 조건을 만족할 때까지 아래로 이동시키는 알고리즘입니다.

### 5. Heap과 알고리즘의 관계

> **"Heap은 우선순위 기반 알고리즘의 핵심 인프라이다"**
>
- **상호작용:** Heap은 데이터를 우선순위 기준으로 정렬된 형태로 유지하고 알고리즘은 이 구조를 활용하여 최적의 값을 빠르게 선택합니다.
- **비유:** Heap은 “항상 가장 중요한 데이터가 위에 올라오도록 정리된 작업 대기열”이고 알고리즘은 “그중 가장 중요한 작업을 선택하여 실행하는 프로세스”입니다.
- **의존성:** 다익스트라, 프림 알고리즘과 같은 최적 경로 탐색 알고리즘은 Heap이 없으면 효율적으로 구현하기 어렵습니다.

### 6. 시간 복잡도 (Heap 기준)

Heap의 연산은 트리의 높이에 의해 결정되며 완전 이진 트리 구조이므로 높이는 log n으로 제한됩니다.

| 연산 | 시간 복잡도 | 핵심 과정 | 특징 설명 |
| :--- | :--- | :--- | :--- |
| **Peek** | $O(1)$ | 루트 노드 접근 | 최댓값 또는 최솟값 조회 |
| **Insert** | $O(\log n)$ | **Heapify Up** | 삽입 후 위로 이동 |
| **Delete** | $O(\log n)$ | **Heapify Down** | 삭제 후 아래로 이 |
| **Build Heap** | $O(n)$ | 상향식 재구성 | 전체 배열을 Heap으로 변환 |

### 최소 힙에서 가장 작은 값 우선 추출

```python
%%writefile min_heap.js

class MinHeap {
    /* ---  Heap 내부 저장소 초기화 --- */
    // Heap 객체가 생성될 때 내부 데이터를 저장할 배열을 초기화 하는 부분
    constructor() {
        this.heap = [];
    }

    /* --- 데이터 삽입 --- */
    // Heap에 새로운 데이터를 삽입하는 함수
    // push()를 이용해 배열의 마지막 위치에 값을 추가한 뒤 heapifyUp()을 호출하여 Heap 조건이 유지되도록 함.
    insert(value) {
        this.heap.push(value);
        this.heapifyUp();
    }

    /* --- Heapify Up 동작 --- */
    // 데이터 삽입 후 수행되는 정렬 복원 알고리즘
    // 부모 노드와 비교하여 부모 값이 더 크면 서로 교환하고 그 위치를 기준으로 비교 동작 반복
    heapifyUp() {
        let i = this.heap.length - 1;
        while (i > 0) {
            // 부모 인덱스 계산
            let parent = Math.floor((i - 1) / 2);
            if (this.heap[parent] <= this.heap[i]) break;
            [this.heap[parent], this.heap[i]] = [this.heap[i], this.heap[parent]];
            i = parent;
        }
    }

    /* --- 최소값 추출 --- */
    // 최소 힙에서 가장 중요한 값인 루트 노드를 꺼내는 함수
    // 최소값(루트 노드)을 추출한 뒤 마지막 노드를 루트 자리로 이동 시켜 heapifyDown()을 수행하며 무너진 최소힙 구조를 복원
    extractMin() {
        if (this.heap.length === 0) return null;
        if (this.heap.length === 1) return this.heap.pop();

        const min = this.heap[0];
        this.heap[0] = this.heap.pop();
        this.heapifyDown();
        return min;
    }

    /* --- Heapify Down 동작 --- */
    // 최소 노드(루트 노드) 삭제 후 수행되는 정렬 복원 알고리즘
    // 마지막 노드가 루트 자리에 들어가면 힙 조건이 깨지기 때문에 이를 복원하기 위해 해당 알고리즘을 수행
    // 왼쪽 자식과 오른쪽 자식 중 더 작은 자식과 비교하여 부모가 더 크면 서로 교환하고 그 위치를 기준으로 비교 동작 반복
    heapifyDown() {
        let i = 0;
        while (2 * i + 1 < this.heap.length) {
            let left = 2 * i + 1;
            let right = 2 * i + 2;
            let smaller = left;

            if (right < this.heap.length && this.heap[right] < this.heap[left]) {
                smaller = right;
            }

            if (this.heap[i] <= this.heap[smaller]) break;

            [this.heap[i], this.heap[smaller]] = [this.heap[smaller], this.heap[i]];
            i = smaller;
        }
    }
}

const heap = new MinHeap();
heap.insert(40);
heap.insert(10);
heap.insert(30);
heap.insert(5);

console.log(heap.extractMin()); // 5
console.log(heap.extractMin()); // 10
```

### 최대 힙에서 가장 큰 값 우선 추출

```python
%%writefile max_heap.js

class MaxHeap {
    constructor() {
        this.heap = [];
    }

    insert(value) {
        this.heap.push(value);
        this.heapifyUp();
    }

    heapifyUp() {
        let i = this.heap.length - 1;
        while (i > 0) {
            let parent = Math.floor((i - 1) / 2);
            if (this.heap[parent] >= this.heap[i]) break;
            [this.heap[parent], this.heap[i]] = [this.heap[i], this.heap[parent]];
            i = parent;
        }
    }

    extractMax() {
        if (this.heap.length === 0) return null;
        if (this.heap.length === 1) return this.heap.pop();

        const max = this.heap[0];
        this.heap[0] = this.heap.pop();
        this.heapifyDown();
        return max;
    }

    heapifyDown() {
        let i = 0;
        while (2 * i + 1 < this.heap.length) {
            let left = 2 * i + 1;
            let right = 2 * i + 2;
            let larger = left;

            if (right < this.heap.length && this.heap[right] > this.heap[left]) {
                larger = right;
            }

            if (this.heap[i] >= this.heap[larger]) break;

            [this.heap[i], this.heap[larger]] = [this.heap[larger], this.heap[i]];
            i = larger;
        }
    }
}

const heap = new MaxHeap();
heap.insert(40);
heap.insert(10);
heap.insert(70);
heap.insert(25);

console.log(heap.extractMax()); // 70
console.log(heap.extractMax()); // 40
```

### 우선순위 큐 구현

```python
%%writefile priority_queue.js

class PriorityQueue {
    constructor() {
        this.heap = [];
    }

    enqueue(value, priority) {
        this.heap.push({ value, priority });
        this.heapifyUp();
    }

    heapifyUp() {
        let i = this.heap.length - 1;
        while (i > 0) {
            let parent = Math.floor((i - 1) / 2);
            if (this.heap[parent].priority <= this.heap[i].priority) break;
            [this.heap[parent], this.heap[i]] = [this.heap[i], this.heap[parent]];
            i = parent;
        }
    }

    dequeue() {
        if (this.heap.length === 0) return null;
        if (this.heap.length === 1) return this.heap.pop();

        const top = this.heap[0];
        this.heap[0] = this.heap.pop();
        this.heapifyDown();
        return top;
    }

    heapifyDown() {
        let i = 0;
        while (2 * i + 1 < this.heap.length) {
            let left = 2 * i + 1;
            let right = 2 * i + 2;
            let smaller = left;

            if (right < this.heap.length && this.heap[right].priority < this.heap[left].priority) {
                smaller = right;
            }

            if (this.heap[i].priority <= this.heap[smaller].priority) break;

            [this.heap[i], this.heap[smaller]] = [this.heap[smaller], this.heap[i]];
            i = smaller;
        }
    }
}

const pq = new PriorityQueue();
pq.enqueue("긴급 장애 처리", 1);
pq.enqueue("일반 문의", 5);
pq.enqueue("보안 점검", 2);

console.log(pq.dequeue()); // { value: '긴급 장애 처리', priority: 1 }
console.log(pq.dequeue()); // { value: '보안 점검', priority: 2 }
```

### 배열을 Heap으로 변환하기

```python
%%writefile arr_to_heap.js

function heapify(arr, n, i) {
    let smallest = i;
    let left = 2 * i + 1;
    let right = 2 * i + 2;

    if (left < n && arr[left] < arr[smallest]) {
        smallest = left;
    }

    if (right < n && arr[right] < arr[smallest]) {
        smallest = right;
    }

    if (smallest !== i) {
        [arr[i], arr[smallest]] = [arr[smallest], arr[i]];
        heapify(arr, n, smallest);
    }
}

function buildMinHeap(arr) {
    const n = arr.length;
    for (let i = Math.floor(n / 2) - 1; i >= 0; i--) {
        heapify(arr, n, i);
    }
    return arr;
}

const arr = [30, 20, 50, 10, 40, 5];
console.log(buildMinHeap(arr)); // [5, 10, 30, 20, 40, 50] 형태의 최소 힙 구조
```

### Heap Sort 구현

```python
%%writefile heap_sort.js

function heapify(arr, n, i) {
    let largest = i;
    let left = 2 * i + 1;
    let right = 2 * i + 2;

    if (left < n && arr[left] > arr[largest]) {
        largest = left;
    }

    if (right < n && arr[right] > arr[largest]) {
        largest = right;
    }

    if (largest !== i) {
        [arr[i], arr[largest]] = [arr[largest], arr[i]];
        heapify(arr, n, largest);
    }
}

function heapSort(arr) {
    const n = arr.length;

    for (let i = Math.floor(n / 2) - 1; i >= 0; i--) {
        heapify(arr, n, i);
    }

    for (let i = n - 1; i > 0; i--) {
        [arr[0], arr[i]] = [arr[i], arr[0]];
        heapify(arr, i, 0);
    }

    return arr;
}

console.log(heapSort([40, 10, 30, 50, 20])); // [10, 20, 30, 40, 50]
```

### Top-K 큰 값 찾기

```python
%%writefile top_k.js

class MinHeap {
    constructor() {
        this.heap = [];
    }

    insert(value) {
        this.heap.push(value);
        this.heapifyUp();
    }

    heapifyUp() {
        let i = this.heap.length - 1;
        while (i > 0) {
            let parent = Math.floor((i - 1) / 2);
            if (this.heap[parent] <= this.heap[i]) break;
            [this.heap[parent], this.heap[i]] = [this.heap[i], this.heap[parent]];
            i = parent;
        }
    }

    extractMin() {
        if (this.heap.length === 1) return this.heap.pop();
        const min = this.heap[0];
        this.heap[0] = this.heap.pop();
        this.heapifyDown();
        return min;
    }

    heapifyDown() {
        let i = 0;
        while (2 * i + 1 < this.heap.length) {
            let left = 2 * i + 1;
            let right = 2 * i + 2;
            let smaller = left;

            if (right < this.heap.length && this.heap[right] < this.heap[left]) {
                smaller = right;
            }

            if (this.heap[i] <= this.heap[smaller]) break;

            [this.heap[i], this.heap[smaller]] = [this.heap[smaller], this.heap[i]];
            i = smaller;
        }
    }

    peek() {
        return this.heap[0];
    }

    size() {
        return this.heap.length;
    }
}

function topK(nums, k) {
    const heap = new MinHeap();

    for (const num of nums) {
        if (heap.size() < k) {
            heap.insert(num);
        } else if (num > heap.peek()) {
            heap.extractMin();
            heap.insert(num);
        }
    }

    return heap.heap;
}

console.log(topK([3, 1, 10, 5, 20, 8, 15], 3)); // 상위 3개 값 유지
```

### 작업 스케줄링 시뮬레이션

```python
%%writefile task_schedule.js

class TaskQueue {
    constructor() {
        this.heap = [];
    }

    addTask(name, priority) {
        this.heap.push({ name, priority });
        this.heapifyUp();
    }

    heapifyUp() {
        let i = this.heap.length - 1;
        while (i > 0) {
            let parent = Math.floor((i - 1) / 2);
            if (this.heap[parent].priority <= this.heap[i].priority) break;
            [this.heap[parent], this.heap[i]] = [this.heap[i], this.heap[parent]];
            i = parent;
        }
    }

    runTask() {
        if (this.heap.length === 0) return null;
        if (this.heap.length === 1) return this.heap.pop();

        const top = this.heap[0];
        this.heap[0] = this.heap.pop();
        this.heapifyDown();
        return top;
    }

    heapifyDown() {
        let i = 0;
        while (2 * i + 1 < this.heap.length) {
            let left = 2 * i + 1;
            let right = 2 * i + 2;
            let smaller = left;

            if (right < this.heap.length && this.heap[right].priority < this.heap[left].priority) {
                smaller = right;
            }

            if (this.heap[i].priority <= this.heap[smaller].priority) break;

            [this.heap[i], this.heap[smaller]] = [this.heap[smaller], this.heap[i]];
            i = smaller;
        }
    }
}

const scheduler = new TaskQueue();
scheduler.addTask("백업", 3);
scheduler.addTask("긴급 패치", 1);
scheduler.addTask("로그 정리", 5);

console.log(scheduler.runTask()); // 긴급 패치
console.log(scheduler.runTask()); // 백업
console.log(scheduler.runTask()); // 로그 정리
```

### 환자 응급도 처리 시스템

```python
%%writefile emergency_queue.js

class EmergencyQueue {
    constructor() {
        this.heap = [];
    }

    addPatient(name, level) {
        this.heap.push({ name, level });
        this.heapifyUp();
    }

    heapifyUp() {
        let i = this.heap.length - 1;
        while (i > 0) {
            let parent = Math.floor((i - 1) / 2);
            if (this.heap[parent].level <= this.heap[i].level) break;
            [this.heap[parent], this.heap[i]] = [this.heap[i], this.heap[parent]];
            i = parent;
        }
    }

    treatPatient() {
        if (this.heap.length === 0) return null;
        if (this.heap.length === 1) return this.heap.pop();

        const patient = this.heap[0];
        this.heap[0] = this.heap.pop();
        this.heapifyDown();
        return patient;
    }

    heapifyDown() {
        let i = 0;
        while (2 * i + 1 < this.heap.length) {
            let left = 2 * i + 1;
            let right = 2 * i + 2;
            let smaller = left;

            if (right < this.heap.length && this.heap[right].level < this.heap[left].level) {
                smaller = right;
            }

            if (this.heap[i].level <= this.heap[smaller].level) break;

            [this.heap[i], this.heap[smaller]] = [this.heap[smaller], this.heap[i]];
            i = smaller;
        }
    }
}

const er = new EmergencyQueue();
er.addPatient("환자A", 3);
er.addPatient("환자B", 1);
er.addPatient("환자C", 2);

console.log(er.treatPatient()); // 응급도 1인 환자B 먼저 처리
```

# 트라이 (Trie)

### 1. Trie의 정의와 목적

**정의**

Trie(트라이)는 문자열을 효율적으로 저장하고 탐색하기 위해 설계된 트리 기반 자료구조로서 각 노드가 하나의 문자를 나타내고 루트부터 특정 노드까지의 경로가 하나의 문자열을 의미하는 구조이며 공통 접두어(prefix)를 공유하는 문자열들을 하나의 경로로 압축하여 저장하는 특징을 가지며 검색 시 문자열을 문자 단위로 분해하여 트리를 따라가면서 탐색하는 방식으로 동작한다.

**목적**

- **문자열 탐색 최적화:** 문자열 길이를 기준으로 O(L) 시간에 검색 가능하도록 한다.
- **접두어 기반 검색:** 자동완성, 추천 시스템과 같이 prefix 기반 탐색을 효율적으로 처리한다.
- **중복 데이터 제거:** 공통 접두어를 공유하여 메모리를 절약한다.
- **빠른 존재 여부 확인:** 문자열 존재 여부를 빠르게 판단할 수 있다.

### 2. Trie의 구조와 메모리 표현

Trie는 트리 구조이며 각 노드는 자식 노드를 참조하는 구조로 구성된다.

- 각 노드는 `children` 객체를 통해 다음 문자로 연결된다.
- 문자열의 끝은 `isEnd` 플래그로 구분된다.
- 실제 메모리에서는 객체(Reference Type) 기반으로 Heap 영역에 저장된다.

### 3. 핵심 알고리즘

**Insert**

문자열을 한 글자씩 순회하면서 존재하지 않는 노드를 생성하고 마지막 문자에 도달하면 종료 표시를 한다.

**Search**

문자열을 한 글자씩 따라가며 경로가 존재하는지 확인하고 마지막 노드의 종료 여부를 확인한다.

**Prefix Search**

문자열 전체가 아닌 특정 접두어까지만 탐색하여 해당 prefix로 시작하는 문자열이 존재하는지 확인한다.

### 4. Trie와 알고리즘의 관계

> **"Trie는 문자열 탐색 알고리즘의 핵심 인프라이다"**
>
- **상호작용:** Trie는 문자열을 구조적으로 저장하고 알고리즘은 이를 기반으로 빠른 탐색 및 추천을 수행한다.
- **비유:** Trie는 “단어 사전의 색인 구조”이고 알고리즘은 “해당 색인을 따라가며 단어를 찾는 과정”이다.
- **의존성:** 자동완성, 사전 검색, 문자열 필터링 알고리즘은 Trie 구조를 기반으로 성능이 결정된다.

### 5. 시간 복잡도 (Trie 기준)

Trie의 연산은 문자열 길이 L에 의해 결정되며 데이터 개수와 무관하게 일정한 탐색 성능${O(L)}$을 가진다.

| 연산 | 시간 복잡도 | 특징 | 설명 |
| :--- | :--- | :--- | :--- |
| **Insert** | $O(L)$ | 문자 단위 삽입 | 문자열 길이만큼 노드 생성 |
| **Search** | $O(L)$ | 문자 단위 탐색 | 경로 존재 여부 확인 |
| **StartsWith** | $O(L)$ | 접두어 탐색 | prefix 기반 검색 |
| **Delete** | $O(L)$ | 조건적 삭제 | 불필요 노드 제거 |

```python
%%writefile trie_node.js

class TrieNode {
    constructor() {
        this.children = {};
        this.isEnd = false;
    }
}

class Trie {
    constructor() {
        this.root = new TrieNode();
    }

    insert(word) {
        let node = this.root;
        for (const char of word) {
            if (!node.children[char]) {
                node.children[char] = new TrieNode();
            }
            node = node.children[char];
        }
        node.isEnd = true;
    }

    search(word) {
        let node = this.root;
        for (const char of word) {
            if (!node.children[char]) return false;
            node = node.children[char];
        }
        return node.isEnd;
    }

    startsWith(prefix) {
        let node = this.root;
        for (const char of prefix) {
            if (!node.children[char]) return false;
            node = node.children[char];
        }
        return true;
    }
}

const trie = new Trie();
trie.insert("apple");
trie.insert("app");

console.log(trie.search("apple"));   // true
console.log(trie.search("app"));     // true
console.log(trie.search("appl"));    // false
console.log(trie.startsWith("app")); // true
```

## Trie 주요 예시

```python
%%writefile trie_practice.js

class TrieNode {
  constructor() {
    this.children = {};
    this.isEnd = false;
  }
}

class Trie {
  constructor() {
    this.root = new TrieNode();
  }

  insert(word) {
    let node = this.root;
    for (const char of word) {
      if (!node.children[char]) {
        node.children[char] = new TrieNode();
      }
      node = node.children[char];
    }
    node.isEnd = true;
  }

  search(word) {
        let node = this.root;
        for (const char of word) {
            if (!node.children[char]) return false;
            node = node.children[char];
        }
        return node.isEnd;
  }

  startsWith(prefix) {
        let node = this.root;
        for (const char of prefix) {
            if (!node.children[char]) return false;
            node = node.children[char];
        }
        return true;
  }
}

// 1. 단어 삽입 및 검색
const trie1 = new Trie();
trie1.insert("cat");
trie1.insert("car");

console.log("--- 단어 삽입 및 검색 ---");
console.log(trie1.search("cat")); // true
console.log(trie1.search("cap")); // false

// 2. 접두어(prefix) 검색
const trie2 = new Trie();
trie2.insert("apple");
trie2.insert("app");

console.log("--- 접두어(prefix) 검색 ---");
console.log(trie2.startsWith("app")); // true
console.log(trie2.startsWith("apl")); // false

// 3. 자동완성 기능 구현
function getWords(node, prefix, result) {
    if (node.isEnd) result.push(prefix);
    for (const char in node.children) {
        getWords(node.children[char], prefix + char, result);
    }
}

function autoComplete(trie, prefix) {
    let node = trie.root;
    for (const char of prefix) {
        if (!node.children[char]) return [];
        node = node.children[char];
    }

    const result = [];
    getWords(node, prefix, result);
    return result;
}

const trie3 = new Trie();
trie3.insert("apple");
trie3.insert("app");
trie3.insert("apex");
trie3.insert("감자볶음");
trie3.insert("감자전");
trie3.insert("감자빵");
trie3.insert("감자튀김");

console.log("--- 자동완성 기능 구현 ---");
console.log(autoComplete(trie3, "ap")); // ['apple', 'app', 'apex']
console.log(autoComplete(trie3, "감자"));

// 4. 단어 개수 세기
function countWords(node) {
  let count = node.isEnd ? 1 : 0;
  for (const char in node.children) {
    count += countWords(node.children[char]);
  }
  return count;
}

const trie4 = new Trie();
trie4.insert("a");
trie4.insert("ab");
trie4.insert("abc");

console.log("--- 단어 개수 세기 ---");
console.log(countWords(trie4.root));

// 5. 특정 prefix 개수 세기
function countPrefix(trie, prefix) {
  let node = trie.root;
  for (const char of prefix) {
    if (!node.children[char]) return 0;
    node = node.children[char];
  }
  return countWords(node);
}

const trie5 = new Trie();
trie5.insert("apple");
trie5.insert("app");
trie5.insert("apex");

console.log("--- 특정 prefix 개수 세기 ---");
console.log(countPrefix(trie5, "ape"));

// 6. 가장 긴 접두어 찾기
function longestPrefix(trie, word) {
  let node = trie.root;
  let prefix = "";

  for (const char of word) {
    if (!node.children[char]) break;
    prefix += char;
    node = node.children[char];
  }

  return prefix;
}

const trie6 = new Trie();
trie6.insert("apple");
trie6.insert("app");

console.log("--- 가장 긴 접두어 찾기 ---");
console.log(longestPrefix(trie6, "applesauce"));

// 7. 단어 삭제 (기본 구조)
function deleteWord(node, word, depth = 0) {
  if (!node) return false;

  if (depth === word.length) {
    if (!node.isEnd) return false;
    node.isEnd = false;
    return Object.keys(node.children).length === 0;
  }

  const char = word[depth];
  if (deleteWord(node.children[char], word, depth + 1)) {
    delete node.children[char];
    return Object.keys(node.children).length === 0;
  }

  return false;

}

const trie7 = new Trie();

console.log("--- 단어 삭제 (기본 구조) ---");
trie7.insert("cat");
console.log(`삭제 전: ${trie7.search("cat")}`);

deleteWord(trie7.root, "cat");
console.log(`삭제 후: ${trie7.search("cat")}`);

// 8. 금칙어 필터링
function filterWord(trie, text) {
  for (let i = 0; i < text.length; i++) {
    let node = trie.root;
    let j = i;

    while (j < text.length && node.children[text[j]]) {
      node = node.children[text[j]];
      if (node.isEnd) return true;
      j++;
    }
  }
  return false;
}

const trie8 = new Trie();
trie8.insert("bad");

console.log("--- 금칙어 필터링 ---");
console.log(filterWord(trie8, "this is a bad word")); // true

// 9. 전화번호 접두어 검사
function hasPrefixConflict(numbers) {
    const trie = new Trie();

    for (const num of numbers) {
        let node = trie.root;

        for (const char of num) {
            if (node.isEnd) return true;
            if (!node.children[char]) {
                node.children[char] = new TrieNode();
            }
            node = node.children[char];
        }

        if (Object.keys(node.children).length > 0) return true;
        node.isEnd = true;
    }

    return false;
}

console.log("--- 전화번호 접두어 검사 ---");
console.log(hasPrefixConflict(["123", "456"])); // false
console.log(hasPrefixConflict(["119", "1192"])); // true

// 10. 문자열 사전 정렬 출력
function printSorted(node, prefix = "") {
    if (node.isEnd) console.log(prefix);

    const keys = Object.keys(node.children).sort();
    for (const char of keys) {
        printSorted(node.children[char], prefix + char);
    }
}

const trie10 = new Trie();
trie10.insert("banana");
trie10.insert("apple");
trie10.insert("app");

console.log("--- 문자열 사전 정렬 출력 ---");
printSorted(trie10.root);
// app
// apple
// banana
```
