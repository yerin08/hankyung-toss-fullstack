# 연결 자료구조

## 연결 리스트 (Linked List)

- **추상 자료형(ADT)의 물리적 구현:** 논리적으로는 일렬로 나열된 선형 구조이지만, 실제 메모리(RAM)상에서는 여기저기 흩어져 있는 노드들을 '참조(Reference)'로 엮어 만든 동적 자료구조입니다.
- **노드의 구성 요소:**
    - **Data:** 저장하려는 실제 값 (숫자, 문자열, 객체 등).
    - **Pointer (Next):** 다음 노드의 메모리 주소를 가리키는 이정표. 마지막 노드의 포인터는 `null`을 가리켜 리스트의 끝임을 알립니다.
- **동적 확장성:** 배열처럼 미리 크기를 예약(Capacity)할 필요가 없어, 메모리가 허용하는 한 무한히 확장 가능하며 '메모리 단편화' 문제에서 상대적으로 자유롭습니다.

### 1) 단일 연결 리스트 (Singly Linked List)

- **특징:** 한 방향으로만 흐르는 편도 통행 구조입니다.
- **구현 핵심:** 각 노드는 `next` 속성만 가집니다. `Head` 노드를 놓치면 전체 리스트를 유실하므로 관리가 엄격해야 합니다.
- **장점:** 메모리 오버헤드가 적고 구현이 단순합니다.
- **단점:** 특정 노드의 이전 노드를 찾으려면 다시 `Head`부터 탐색해야 하는 비효율이 발생합니다 ($O(n)$).

### 2) 이중 연결 리스트 (Doubly Linked List)

- **특징:** 양방향 통행이 가능한 구조입니다.
- **구현 핵심:** 노드가 `prev`와 `next` 두 개의 포인터를 가집니다.
- **장점:** 특정 노드 삭제 시 이전 노드를 즉시 알 수 있어 삭제 작업이 매우 유연하며, 뒤에서부터 역방향 탐색이 가능합니다.
- **단점:** 포인터 변수를 하나 더 저장해야 하므로 노드당 메모리 사용량이 약 1.5~2배 증가하며, 삽입/삭제 시 변경해야 할 포인터가 많아 코드가 복잡해집니다.

## **예시 코드**

### 기초 구조 및 단일 연결 리스트

```python
%%writefile ex01.js

// 1) 노드 생성 함수 및 리스트 초기화

// [단계 1] 노드 생성 함수 정의
const createNode = (data) => ({
  data,
  next: null
});

// [단계 2] 리스트 초기 상태 설정
let list = {
  head: null,
  size: 0
};

// [단계 3] 초기 상태 출력 확인
console.log("--- [초기화 상태] ---");
console.log("리스트 객체:", list);
console.log("Head 연결 여부:", list.head === null ? "비어 있음" : "연결됨");
console.log("현재 크기:", list.size);

// 2) 맨 앞에 노드 삽입 (prepend) - O(1)

// [실행 전] 초기 상태: { head: null, size: 0 }

function prepend(list, data) {
  const newNode = createNode(data);
  newNode.next = list.head;
  list.head = newNode;
  list.size++;

  console.log(`[prepend] '${data}' 삽입 후 리스트 상태:`, JSON.stringify(list, null, 2));
}

// 실행 예시 및 출력
prepend(list, "데이터1");
// 출력: { "head": { "data": "데이터1", "next": null }, "size": 1 }

prepend(list, "데이터2");
// 출력: { "head": { "data": "데이터2", "next": { "data": "데이터1", "next": null }}, "size": 2 }

// 3) 맨 뒤에 노드 삽입 (append) - O(n)

// [실행 전] 초기 상태: { head: { data: '데이터2', next: { data: '데이터1', next: null } }, size: 2 }
function append(list, data) {
  const newNode = createNode(data);
  if (!list.head) {
    list.head = newNode;
  } else {
    let current = list.head;
    while (current.next) {
      current = current.next;
    }
    current.next = newNode;
  }
  list.size++;

  console.log(`[append] '${data}' 삽입 후 리스트 상태:`, JSON.stringify(list, null, 2));
}

// 실행 예시 및 출력
append(list, "데이터3");
/* 출력: {
  "head": {
    "data": "데이터2",
    "next": {
      "data": "데이터1",
      "next": {
        "data": "데이터3",
        "next": null
      }
    }
  },
  "size": 3
}
*/

// 4) 특정 값 삭제 (remove) - O(n)

// [실행 전] 초기 상태: { "head": { "data": "데이터2", "next": { "data": "데이터1", "next": { "data": "데이터3", "next": null } } }, "size": 3 }

function remove(list, data) {
  if (!list.head) return;

  if (list.head.data === data) {
    list.head =list.head.next;
    list.size--;
    console.log(`[remove] Head 노드 '${data}' 삭제 후 리스트 상태:`, JSON.stringify(list, null, 2));
    return;
  }

  let curr = list.head;
  while (curr.next && curr.next.data !== data) { curr = curr.next; }

  if (curr.next) {
    curr.next = curr.next.next;
    list.size--;
    console.log(`[remove] 중간/끝 노드 '${data}' 삭제 후 리스트 상태:`, JSON.stringify(list, null, 2));
  } else {
    console.log(`[remove 실패] '${data}'를 찾을 수 없습니다.`);
  }
}

// 실행 예시 및 출력
remove(list, "데이터1");
/* 출력: {
  "head": {
    "data": "데이터2",
    "next": {
      "data": "데이터3",
      "next": null
    }
  },
  "size": 2
}
*/
```

### 이중 연결 리스트 (Doubly Linked List)

```python
%%writefile ex02.js

// 5) 이중 연결 노드 생성

// [실행 전] 이중 연결 리스트 초기 리스트: { head: null, tail: null, size: 0 }

const createDoublyNode = (data) => ({
  data,
  prev: null,
  next: null
});

let list = {
  head: null,
  size: 0
};

// 이중 연결 리스트 삽입 예시 (가장 앞)
function prependDoubly(list, data) {
  const newNode = createDoublyNode(data);

  if (!list.head) {
    list.head = newNode;
    list.tail = newNode;
  } else {
    newNode.next = list.head;
    list.head.prev = newNode;
    list.head = newNode;
  }
  list.size++;

  console.log(`[Doubly-Prepend] '${data}' 삽입 후 상태:`, JSON.stringify(list, null, 2));
}

// 실행 예시 및 출력
prependDoubly(list, "이중데이터1");
/* 출력: {
  "head": {
    "data": "이중데이터1",
    "prev": null,
    "next": null
  },
  "tail": { ... 동일 노드 참조 ... },
  "size": 1
}
*/

// 6) 중간 삽입 로직 (포인터 4개 조작)

// [실행 전] 상황: nodeA <-> nodeB (연결된 상태)

function insertBetween(nodeA, nodeB, data) {
  const newNode = createDoublyNode(data);

  // A와 B 사이에 NEW를 넣는 핵심 로직
  newNode.next = nodeB;
  newNode.prev = nodeA;
  nodeA.next = newNode;
  nodeB.prev = newNode;

  console.log(`[InsertBetween] '${nodeA.data}와 '${nodeB.data}' 사이에 '${data}' 삽입 완료`);
  console.log(`결과 구조: ${nodeA.data} <-> ${newNode.data} <-> ${nodeB.data}`);
}

// 실행 예시 및 출력
// (nodeA : "데이터1", nodeB: "데이터2"가 있다고 가정)
const nodeA = createDoublyNode("데이터1");
const nodeB = createDoublyNode("데이터2");
insertBetween(nodeA, nodeB, "NEW");

/* 출력:
[InsertBetween] '데이터1'와 '데이터2' 사이에 'NEW' 삽입 완료
결과 구조: 데이터1 <-> NEW <-> 데이터2
*/
```

```python
%%writefile ex03.js

const createNode = (data) => ({
  data,
  next: null
});

function prepend(list, data) {
  const newNode = createNode(data);
  newNode.next = list.head;
  list.head = newNode;
  list.size++;
}

/* ----- 배열 vs 연결 리스트 성능 비교 실험 ----- */

// 7) 배열의 맨 앞 삽입 성능 (O(n))

// [배열의 성능 측정] 데이터 10만 개를 맨 앞에 삽입할 때

const arr = [];

console.time('Array unshift'); // 타이머 시작

for (let i = 0; i < 100000; i++) {
  arr.unshift(i); // 기존 인덱스 10만 개를 매번 뒤로 밀어냄 (O(n))
}

console.timeEnd('Array unshift'); // 타이머 종료 및 소요 시간 출력

/* 출력 예시 (시스템 환경에 따라 다름):
Array unshift: 2542.123ms
- 설명: 배열은 새로운 요소를 맨 앞에 넣을 때마다 기존의 모든 요소(최대 10만 개)를
  메모리상에서 한 칸씩 뒤로 이동시켜야 하므로 데이터가 많아질수록 기하급수적으로 느려집니다.
*/

// 8) 연결 리스트의 맨 앞 삽입 성능 (O(1))

// [연결 리스트의 성능 측정] 데이터 10만 개를 맨 앞에 삽입할 때

let myList = { head: null };

console.time('List prepend'); // 타이머 시작

for (let i = 0; i < 100000; i++) {
  prepend(myList, i); // 참조값 1개(head)만 바꾸고 끝냄
}

console.timeEnd('List prepend'); // 타이머 종료 및 소요 시간 출력

/* 출력 예시 (시스템 환경에 따라 다름):
List prepend: 3.456ms
- 설명: 연결 리스트는 데이터가 10만 개든 1억 개든,
  단순히 새 노드의 next를 기존 head에 연결하고 head를 교체하는 작업만 수행하므로
  배열(unshift)에 비해 압도적으로 빠른 성능을 보여줍니다.
*/
```

### 실전 응용 및 기타

```python
%%writefile ex04.js

// 9) 연결 리스트 기반 큐(Queue)의 Dequeue

let list = {
  head: { data: "데이터2", next: { data: "데이터 3", next: null } },
  size: 2
};

// [실행 전] 초기 상태: { "head": { "data": "데이터2", "next": { "data": "데이터3", "next": null } }, "size": 2 }

function dequeue(list) {
  if (!list.head) return null;

  const removedData = list.head.data;
  list.head = list.head.next;
  list.size--;

  console.log(`[dequeue] '${removedData}' 추출 완료. 새로운 Head: ${list.head ? list.head.data : 'null'}`);
  return removedData;
}

// 실행 예시 및 출력
const data = dequeue(list);
/* 출력:
[dequeue] '데이터2' 추출 완료. 새로운 Head: 데이터3
최종 리스트 상태: { "head": { "data": "데이터3", "next": null }, "size": 1 }
- 설명: 연결 리스트를 활용한 큐(Queue)에서 dequeue는
  단순히 head의 참조만 다음으로 옮기면 되므로 O(1)의 성능을 유지합니다.
*/

// 10) 순환(Circular) 구조 만들기

// [실행 전] 상황: Head -> [데이터3] -> null

function makeCircular(list) {
  if (!list.head) return;

  let last = list.head;

  // 1. 마지막 노드(Tail)를 찾을 때까지 이동
  while (last.next) {
    last = last.next;
  }

  // 2. 마지막 노드의 next를 다시 head로 연결 (순환 고리 형성)
  last.next = list.head;

  console.log(`[Circular] 마지막 노드 '${last.data}'의 next를 Head '${list.head.data}'로 연결 완료`);
  console.log(`확인: list.head.next... (무한 반복 가능 구조)`);
}

// 실행 예시 및 출력
makeCircular(list);

/* 출력:
[Circular] 마지막 노드 '데이터3'의 next를 Head '데이터3'로 연결 완료
- 설명: 이제 마지막 노드는 null을 가리키지 않고 다시 처음으로 돌아갑니다.
  이 구조는 '무한 루프 재생 목록'이나 '라운드 로빈 스케줄링'에 활용됩니다.
*/
```

## 연결 리스트 구현

```python
%%writefile list_ex01.js

/* ----- 1) Node 구조 설계 및 리스트 초기화 ----- */

// 노드 생성 함수: 데이터와 next 포인터를 가진 객체 반환
const createNode = (data) => ({
  data: data,
  next: null
});

// 연결 리스트 초기 상태 (Head가 비어있음)
let myLinkedList = { head: null, size: 0 };

console.log("--- 리스트 초기화 완료 ---");
console.log(myLinkedList);

/* ----- 2) 삽입 (Insertion) - 맨 앞과 맨 뒤 ----- */

// 맨 앞에 삽입 (Prepend): O(1)
function insertAtFront(list, data) {
  const newNode = createNode(data);
  newNode.next = list.head;
  list.head = newNode;
  list.size++;
  console.log(`[삽입] 맨 앞에 ${data} 추가 완료`);
}

// 맨 뒤에 삽입 (Append): O(n)
function insertAtBack(list, data) {
  const newNode = createNode(data);
  if (!list.head) {
    list.head = newNode;
  } else {
    let curr = list.head;
    while (curr.next) {
      curr = curr.next;
    }
    curr.next = newNode;
  }
  list.size++;
  console.log(`[삽입] 맨 뒤에 ${data} 추가 완료`);
}

insertAtFront(myLinkedList, "노드1");
insertAtBack(myLinkedList, "노드2");
insertAtBack(myLinkedList, "노드3");

/* ----- 3) 탐색 (Search) - 특정 데이터 찾기 ----- */

function findNode(list, targetData) {
  let curr = list.head;
  let index = 0;

  while(curr) {
    if (curr.data === targetData) {
      console.log(`[탐색 성공] '${targetData}'를 인덱스 ${index}에서 찾았습니다.`);
      return curr;
    }
    curr = curr.next;
    index++;
  }
  console.log(`[탐색 실패] '${targetData}'가 리스트에 없습니다.`);
  return null;
}

findNode(myLinkedList, "노드2");
findNode(myLinkedList, "노드99");

/* ----- 4) 삭제 (Deletion) - 특정 노드 연결 끊기 ----- */

function deleteNode(list, targetData) {
  if (!list.head) return;

  // 1. 삭제할 대상이 Head일 때
  if (list.head.data === targetData) {
    list.head = list.head.next;
    list.size--;
    console.log(`[삭제] Head 노드 '${targetData}' 삭제 완료`);
    return;
  }

  // 2. 삭제할 대상의 이전 노드(prev) 찾기
  let prev = list.head;
  while(prev.next && prev.next.data !== targetData) {
    prev = prev.next;
  }

  // 3. 연결 변경 (건너뛰기)
  if (prev.next) {
    prev.next = prev.next.next;
    list.size--;
    console.log(`[삭제] 중간 노드 '${targetData}' 삭제 완료`);
  } else {
    console.log(`[삭제 실패] '${targetData}'를 찾을 수 없습니다.`);
  }
}

deleteNode(myLinkedList, "노드2");

/* ----- 5) 전체 리스트 출력 (Verification) ----- */

function printList(list) {
  let curr = list.head;
  let result = "Head -> ";

  while(curr) {
    result += `[${curr.data}] -> `;
    curr = curr.next;
  }

  result += "null";
  console.log("--- 현재 리스트 상태 ---");
  console.log(result);
  console.log(`총 노드 개수: ${list.size}`);
}

printList(myLinkedList);
```

### 문제 1. 단일 연결 리스트: 마지막 인원 확인 (Tail 탐색)

- **상황:** 연결 리스트 형태로 줄을 서 있는 명단이 주어집니다. 각 노드는 다음 사람을 가리킵니다. 명단의 가장 마지막에 서 있는 사람의 이름을 찾아내는 함수를 작성하세요.
- **핵심:** `while` 문을 사용하여 `curr.next`가 `null`인 노드(마지막 노드)에 도달할 때까지 포인터를 이동시킵니다.
- **Input:** `head: { data: "철수", next: { data: "영희", next: { data: "민수", next: null } } }`
- **Output:** `"민수"`

```python
%%writefile test01.js

function searchLast(list) {
  let curr = list.head;

  while(curr.next) {
    curr = curr.next;
  }

  return curr.data;
}

const list = {
  head: { data: "철수", next: { data: "영희", next: { data: "민수", next: null } } }
};

console.log(searchLast(list));
```

### 문제 2. 단일 연결 리스트: 중간 새치기 (Node 삽입)

- **상황:** `nodeA`와 `nodeB`가 연결되어 있습니다. 두 사람 사이에 새로운 사람(`newNode`)이 새치기를 하려고 합니다. 연결 관계를 수정하여 `nodeA -> newNode -> nodeB`가 되도록 만드세요.
- **핵심:** `newNode.next`에 기존의 `nodeA.next`를 먼저 연결한 뒤, `nodeA.next`를 `newNode`로 교체하여 연결 고리를 재구성합니다.
- **Input:** `nodeA` (앞사람), `newNode` (새치기객)
- **Output:** `nodeA.next.data`가 `newNode`의 데이터와 일치해야 함

```python
%%writefile test02.js

function addNode(nodeA, newNode) {
  newNode.next = nodeA.next;
  nodeA.next = newNode;
}

const nodeB = { data: "nodeB", next: null };
const nodeA = { data: "nodeA", next: nodeB };
const newNode = { data: "newNode", next: null };

console.log(`newNode 삽입 전: nodeA -> ${nodeA.next.data}`);
addNode(nodeA, newNode);
console.log(`newNode 삽입 후: nodeA ->${nodeA.next.data}`);
```

### 문제 3. 단일 연결 리스트: 명단 합산 (전체 탐색)

- **상황:** 각 노드에 숫자가 적힌 쪽지를 들고 사람들이 서 있습니다. `head`부터 시작하여 마지막 사람까지 이동하며 모든 쪽지에 적힌 숫자의 총합을 구하세요.
- **핵심:** `curr` 변수가 `null`이 될 때까지 반복하며 `curr.data`를 누적 변수에 더하고, `curr = curr.next`로 한 칸씩 전진합니다.
- **Input:** `10 -> 20 -> 30 -> null`
- **Output:** `60`

```python
%%writefile test03.js

function nodeSum(list) {
  let curr = list;
  let sum = 0;

  while(curr) {
    sum += curr.data;
    curr = curr.next;
  }

  return sum;
}

const list = { data: 10, next: { data: 20, next: { data: 30, next: null } } };
console.log(nodeSum(list));
```
