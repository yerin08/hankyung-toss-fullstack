# 스택 (Stack)

### 예제 1: 스택의 기본 동작과 상태 변화 (LIFO 이해)

JavaScript 배열을 스택으로 추상화하여 데이터가 삽입되고 삭제될 때의 내부 상태를 확인하는 가장 기초적인 예제

```python
%%writefile stack_ex01.js

// 스택 예제 1: 기본 Push & Pop
const stack = [];

// 1. 데이터 삽입 (Push)
stack.push('데이터 A');
stack.push('데이터 B');
stack.push('데이터 C');
// 현재 상태: ['데이터 A', '데이터 B', '데이터 C'] (Top: 데이터 C)

// 2. 데이터 추출 (Pop)
const popped =  stack.pop();

console.log(`추출된 요소: ${popped}`);
console.log(`현재 스택 상태: ${stack}`);
```

### 예제 3: 문자열 역순 출력 (String Reversal)

스택의 '후입선출' 특성을 이용하여 입력된 문자열을 거꾸로 뒤집는 알고리즘 예제

```python
%%writefile stack_ex03.js

// 스택 예제 3: 문자열 뒤집기
const text = "HELLO";
const charStack = [];
let reversedText = "";

// 1. 글자를 하나씩 스택에 Push
for (let char of text) {
  charStack.push(char);
}

// 2. 스택에서 하나씩 Pop하여 새로운 문자열 생성
while (charStack.length > 0) {
  reversedText += charStack.pop();
}

console.log(`역순 결과: ${reversedText}`);
console.log(`스택 상태: ${charStack}`);
```

```python
!node stack_ex03.js
```

### 예제 4: 수식의 괄호 유효성 검사 (Parentheses Checker)

여는 괄호가 나오면 스택에 넣고, 닫는 괄호가 나오면 스택에서 짝을 찾아 제거하며 수식의 문법이 맞는지 확인하는 예제

```python
%%writefile stack_ex04.js

// 스택 예제 4: 괄호 짝 맞추기
const expression = "( ( 1 + 2 ) * 3 )";
const parenStack = [];
let isBalanced = true;

for (let char of expression) {
    if (char === '(') {
        parenStack.push(char);
    } else if (char === ')') {
        if (parenStack.length === 0) {
            isBalanced = false;
            break;
        }
        parenStack.pop();
    }
}

if (parenStack.length !== 0) isBalanced = false;

console.log(`수식의 유효성: ${isBalanced}`);
console.log(`최종 스택: ${parenStack}`);
```

```python
!node stack_ex04.js
```

### 예제 4: 연결 리스트 기반 큐 (성능 최적화 구조)

shift()의 성능 문제를 해결하기 위해 클래스를 활용한 효율적인 큐 구조의 예시

```python
%%writefile queue_ex04.js

// 큐 예제 4: 객체를 활용한 효율적 큐 (O(1) 구현)
class SimpleQueue {
  constructor() {
    this.storage = {};
    this.front = 0;
    this.rear = 0;
  }

  enqueue(item) {
    this.storage[this.rear] = item;
    this.rear++;
  }

  dequeue() {
    const item = this.storage[this.front];
    delete this.storage[this.front];
    this.front++;
    return item;
  }
}

const fastQueue = new SimpleQueue();
fastQueue.enqueue('데이터 1');
fastQueue.enqueue('데이터 2');

console.log(`추출 데이터: ${fastQueue.dequeue()}`);
```

```python
!node queue_ex04.js
```

### 문제 1. 스택(Stack): 후순위 인원 관리 (올바른 짝 찾기)

- **상황**: 명단에 `(`와 `)`가 섞여 들어옵니다. `(`는 '입장 예정', `)`는 '입장 완료'를 뜻합니다. 모든 '입장 예정' 인원이 누락 없이 '입장 완료' 되었는지 검사하는 함수를 작성하세요.
- **핵심**: 스택에 `(`를 쌓고 `)`를 만날 때마다 `pop()`하여 짝이 맞는지 확인합니다.
- **Input**: `"(())()"` (정상), `"(()"` (비정상)
- **Output**: `true` 또는 `false`

```python
%%writefile test01.js

function checkEntries(str) {
    const entryStack = [];
    let result = true;

    for (let char of str) {
        if (char === '(') {
            entryStack.push(char);
        } else {
            if (entryStack.length === 0) {
                result = false;
            }
            entryStack.pop();
        }
    }

    if (entryStack.length !== 0) result = false;

    return result;
}

console.log(checkEntries("(())()"));
console.log(checkEntries("(()"));
```

### 문제 2. 큐(Queue): 요세푸스 문제 (순환 대기열)

- **상황**: `['김', '이', '박', '최', '정']` 5명이 원형으로 앉아있습니다. 1번부터 시작하여 **매 3번째 사람**을 명단에서 제외합니다. 마지막까지 남는 사람의 이름을 반환하세요.
- **핵심**: 큐에서 사람을 꺼내(`shift`) 다시 뒤로 넣는(`push`) 과정을 반복하다가, 3번째 차례에만 다시 넣지 않고 제거합니다.
- **Output**: `마지막 생존자: 박`

```python
%%writefile test02.js

function solveJosephus(nameList) {

  while (nameList.length > 1) {
    for (let i = 0; i < 3; i++) {
      if (i === 2) {
        nameList.shift();
        break;
      }
      nameList.push(nameList.shift());
    }
  }
  return nameList;
}

console.log(`마지막 생존자: ${solveJosephus(['김', '이', '박', '최', '정'])}`);
```

### 문제 3. 덱(Deque): 슬라이딩 윈도우 최대 점수

- **상황**: 연속된 3일간의 방문자 수 `[10, 20, 10, 30, 20]`가 주어집니다. 덱을 사용하여 **구간(크기 3)** 이동 시 가장 큰 방문자 수 합계를 구하세요.
- **핵심**: 덱의 양방향 성질을 이용해 앞 요소를 빼고(`shift`) 뒤 요소를 넣으며(`push`) 실시간 합계를 갱신합니다.
- **Output**: `최대 합계: 60` (20, 10, 30 구간

```python
%%writefile test03.js

function maxSum(visitorList) {
  const window = [];
  const sum = [];
  const MAX_SIZE = 3;

  for (let visitor of visitorList) {
      window.push(visitor);

      if (window.length === MAX_SIZE) {
          sum.push(window.reduce((acc, cur) => acc + cur, 0));
          window.shift();
      }
  }
  return Math.max(...sum);
}

console.log(`최대 합계: ${maxSum([10, 20, 10, 30, 20])}`);
```

### 문제 4. 큐(Queue): 우선순위 프린터

- **상황**: 인쇄 대기열에 `[중요도, 이름]` 형태의 데이터가 있습니다. 현재 큐의 맨 앞 문서보다 중요도가 높은 문서가 뒤에 하나라도 있으면, 맨 앞 문서를 꺼내어 다시 맨 뒤로 보냅니다. 특정 인물의 문서가 몇 번째로 인쇄되는지 구하세요.
- **Input**: `[[2, '김'], [1, '이'], [3, '박'], [2, '최']]`, 대상: `'김'`
- **Output**: `3` (박 -> 최 -> 김 순서)

```python
%%writefile test04.js

function priority(printQueue, target) {
  let count = 0;

  while (printQueue.length > 0) {
    const maxPriority = Math.max(...printQueue.map(item => item[0]));

    const current = printQueue.shift();

    if (current[0] < maxPriority) {
      printQueue.push(current);
    } else {
      count++;
      if (current[1] === target) {
        return count;
      }
    }
  }
}

console.log(priority([[2, '김'], [1, '이'], [3, '박'], [2, '최']], '김'));
```

```python
!node test04.js
```

### 문제 5. 스택(Stack): 중복 이름 연속 제거

- **상황**: 이름 배열에서 **연속으로 같은 이름**이 나오면 두 이름을 모두 제거합니다. 제거 후 또 연속되는 이름이 생기면 반복해서 제거합니다. 최종 남은 명단을 반환하세요.
- **Input**: `['김', '이', '이', '김', '박']`
- **Output**: `['박']` (이-이 제거 -> 김-김 제거 -> 박 남음)

```python
%%writefile test05.js

function reduceNames(nameList) {
    const stack = [];

    for (let name of nameList) {
      if (stack.length === 0) stack.push(name);
      else {
        if (stack[stack.length - 1] === name) {
          stack.pop();
        } else {
          stack.push(name);
        }
      }
    }

    return stack;
}

console.log(reduceNames(['김', '이', '이', '김', '박']));
```

```python
!node test05.js
```
