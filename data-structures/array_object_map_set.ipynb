### 1. Array: Cinema Waiting Queue (배열의 추가와 제거)

```python
%%writefile ex01.js

// 배열 연습: 선입선출(FIFO) 구조 이해하기
const waitingQueue = ['Chulsoo', 'Younghee', 'Minsoo'];

// 1. 새로운 관객 'Chulmin'이 대기열 맨 뒤에 도착
waitingQueue.push('Chulmin');

// 2. 가장 앞에 있던 관객 'Chulsoo'가 영화관에 입장
const attendee = waitingQueue.shift();

console.log(`현재 입장객: ${attendee}`);
console.log(`남은 대기열: ${waitingQueue}`);
```

### 2. Array: Product Price Filtering

```python
%%writefile ex02.js

// 배열 연습: 데이터 필터링 및 가공
const productList = [
  { name: 'Keyboard', price: 30000, owner: 'Chulsoo' },
  { name: 'Monitor', price: 200000, owner: 'Minsoo' },
  { name: 'Mouse', price: 15000, owner: 'Younghee' },
  { name: 'Headset', price: 80000, owner: 'Chulmin' }
];

// 가격이 50,000원 이상인 상품의 소유자 이름만 추출
const expensiveItems = productList
  .filter(item => item.price >= 50000)
  .map(item => item.owner);

console.log(expensiveItems);
```

### 3. Object: Election Vote Counter (객체를 활용한 빈도수 계산)

```python
%%writefile ex03.js

// 객체 연습: 빈도수 측정 (해시 맵 패턴)
const voteRecords = ['Chulsoo', 'Minsoo', 'Chulsoo', 'Younghee', 'Minsoo', 'Chulsoo'];
const voteCounter = {};

for (let name of voteRecords) {
  // 이미 객체에 이름이 있다면 1을 더하고, 없다면 1로 시작
  if (voteCounter[name]) {
    voteCounter[name] += 1;
  } else {
    voteCounter[name] = 1;
  }
}

console.log(voteCounter);
```

```python
!node ex03.js
```

### 6. String & Array: Sentence Reverser (문자열 뒤집기)

```python
%%writefile ex06.js

// 문자열 연습: 단어 순서 뒤집기
const originalSentence = 'Minsoo Younghee Chulsoo';

// 1. 공백을 기준으로 단어를 잘라 배열로 만듦
const wordsArray = originalSentence.split(' ');

// 2. 배열의 순서를 거꾸로 뒤집음
const reversedArray = wordsArray.reverse();

// 3. 배열 요소를 다시 공백으로 연결하여 문자열 생성
const finalResult = reversedArray.join(' ');

console.log(`원본: ${originalSentence}`);
console.log(`결과: ${finalResult}`);
```

```python
!node ex06.js
```

### 7. Array (Stack): Undo Function (되돌리기 기능)

```python
%%writefile ex07.js

// 배열 연습: 스택(Stack)을 활용한 작업 취소
const actionHistory = [];

// 1. 사용자가 수행한 작업들을 기록 (Push)
actionHistory.push('Typed: Minsoo');
actionHistory.push('Typed: loves');
actionHistory.push('Typed: JavaScript');

// 2. 'Ctrl + Z' (실행 취소) 수행: 마지막 작업 제거 (Pop)
const lastAction = actionHistory.pop();

console.log(`취소된 작업: ${lastAction}`);
console.log(`남은 기록: ${actionHistory}`);
```

### 8. Array: Score Sorting (데이터 정렬)

```python
%%writefile ex08.js

// 배열 연습: 성적순 정렬 (내림차순)
const studentScores = [
  { name: 'Minsoo', score: 85 },
  { name: 'Chulmin', score: 92 },
  { name: 'Younghee', score: 78 }
];

// 점수(score)를 기준으로 큰 숫자가 앞으로 오게 정렬
studentScores.sort((a, b) => b.score - a.score)

console.log('--- 성적 우수자 명단 ---');
studentScores.forEach((student, index) => {
  console.log(`${index + 1}등: ${student.name} (${student.score}점)`);
});
```

```python
!node ex08.js
```

### 10. Array: Find and Update (데이터 찾기 및 수정)

```python
%%writefile ex10.js

// 배열 연습: 특정 데이터 수정하기
const attendanceBook = [
  { name: 'Minsoo', present: true },
  { name: 'Chulmin', present: false },
  { name: 'Younghee', present: true }
];

// 지각했던 'Chulmin'이 도착하여 출석 상태를 변경함
const targetStudent = attendanceBook.find(student => student.name === 'Chulmin');

if (targetStudent) {
  targetStudent.present = true;
}

console.log('최종 출석부:', attendanceBook);
```

### 문제 1: 중복 없는 명단과 인원수 구하기

**상황:** 이벤트 응모자 명단이 배열로 주어집니다. 중복된 이름을 제거한 실제 응모자 명단(배열)과 총 몇 명인지를 출력하는 함수를 작성하세요.

- **Input:** `['Minsoo', 'Chulmin', 'Minsoo', 'Younghee', 'Chulmin']`
- **Output:** `명단: ['Minsoo', 'Chulmin', 'Younghee'], 인원: 3`

```python
%%writefile test01.js

function unique(arr) {
  const uniqueSet = new Set(arr);

  console.log(`명단: ${[...uniqueSet]}, 인원: ${uniqueSet.size}`);
}

const applicants = ['Minsoo', 'Chulmin', 'Minsoo', 'Younghee', 'Chulmin'];
unique(applicants);
```

### 문제 2: 과일 바구니 개수 세기 (해시 맵)

**상황:** 바구니에 담긴 과일들의 이름이 배열로 주어집니다. 각 과일이 몇 개씩 들어있는지 객체(Object) 형태로 정리하여 반환하는 함수를 작성하세요.

- **Input:** `['apple', 'banana', 'apple', 'orange', 'banana', 'apple']`
- **Output:** `{ apple: 3, banana: 2, orange: 1 }`

```python
%%writefile test02.js

function counter(arr) {
  const count = {};

  for (let item of arr) {
    if (count[item]) {
      count[item] += 1;
    } else {
      count[item] = 1;
    }
  }

  console.log(count);
}

const fruits = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple'];
counter(fruits);
```

```python
!node test02.js
```
