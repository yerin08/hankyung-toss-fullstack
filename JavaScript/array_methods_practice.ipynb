## мҳҲм ң 1: к°ҖкІ©мқҙ 1000мӣҗ мқҙмғҒмқё кіјмқјмқҳ мқҙлҰ„мқ„ лҢҖл¬ёмһҗлЎң ліҖкІҪн•ҳкё°

**лҚ°мқҙн„° мӨҖл№„:**

items л°°м—ҙм—җ кіјмқј мқҙлҰ„, к°ҖкІ© лҚ°мқҙн„° м ҖмһҘн•ңлӢӨ.

**лҚ°мқҙн„° мІҳлҰ¬:**

1. filter н•ЁмҲҳлҘј нҷңмҡ©н•ҙ items л°°м—ҙм—җ м ‘к·јн•ҳм—¬ к°ҖкІ©мқҙ 1000мӣҗ мқҙмғҒмқё кіјмқјл§Ң м„ лі„н•ҳм—¬ мғҲлЎңмҡҙ л°°м—ҙлЎң л§Ңл“Өм–ҙ л°ҳнҷҳн•ңлӢӨ.

2. мң„м—җм„ң л°ҳнҷҳлҗң мғҲлЎңмҡҙ л°°м—ҙм—җ м ‘к·јн•ҳм—¬ map н•ЁмҲҳлЎң кіјмқј мқҙлҰ„мқ„ лҢҖл¬ёмһҗлЎң ліҖкІҪн•ҳкі  к·ё к°’мқ„ мғҲлЎңмҡҙ л°°м—ҙлЎң л§Ңл“Өм–ҙ result ліҖмҲҳм—җ м ҖмһҘн•ңлӢӨ.

3. result л°°м—ҙмқ„ мҪҳмҶ”м—җ м¶ңл Ҙн•ңлӢӨ.

```python
%%writefile ex01.js

const items = [ // кіјмқј мқҙлҰ„, к°ҖкІ© лҚ°мқҙн„°к°Җ нҸ¬н•Ёлҗң л°°м—ҙ
  { name: 'apple', price: 1200 },
  { name: 'pear', price: 800 },
  { name: 'orange', price: 1500 }
];

const result = items
  .filter(item => item.price >= 1000) // items л°°м—ҙм—җ м ‘к·јн•ҳм—¬ к°ҖкІ©мқҙ 1000мӣҗ мқҙмғҒмқё кіјмқјл§Ң м„ лі„н•ҙ мғҲлЎңмҡҙ л°°м—ҙмқ„ л°ҳнҷҳ [ { name: 'apple', price: 1200 }, { name: 'orange', price: 1500 } ]
  .map(item => item.name.toUpperCase()); // filter н•ЁмҲҳмқҳ кІ°кіјлЎң л°ҳнҷҳлҗң л°°м—ҙм—җ лҳҗ н•ңлІҲ м ‘к·јн•ҳм—¬ кіјмқј мқҙлҰ„мқ„ лҢҖл¬ёмһҗлЎң ліҖнҷҳн•ҳлҠ” н•ЁмҲҳ мҲҳн–ү -> мғҲлЎңмҡҙ л°°м—ҙлЎң л§Ңл“Өм–ҙ л°ҳнҷҳ ['APPLE', 'ORANGE']

console.log(result); // мөңмў… л°°м—ҙмқ„ мҪҳмҶ”м—җ м¶ңл Ҙ
```

```python
!node ex01.js
```

## мҳҲм ң 2-1: forEach - м¶ңм„қл¶Җ лӘ…лӢЁ м¶ңл Ҙн•ҳкё°

**лҚ°мқҙн„° мӨҖл№„:**

students л°°м—ҙм—җ н•ҷмғқл“Өмқҳ мқҙлҰ„мқ„ м ҖмһҘн•ңлӢӨ.

**лҚ°мқҙн„° мІҳлҰ¬:**


forEachлҘј мӮ¬мҡ©н•ҳм—¬ students л°°м—ҙмқ„ мҲңнҡҢн•ҳл©° нҷҳмҳҒ л©”мӢңм§ҖлҘј мҪҳмҶ”м—җ м¶ңл Ҙн•ңлӢӨ. мқҙл•Ң forEachмқҳ кІ°кіјлЎң мғҲлЎңмҡҙ л°°м—ҙмқҙ л°ҳнҷҳлҗҳм§Җ м•ҠлҠ”лӢӨ.


```python
%%writefile ex02_1.js

const students = ['лҜјмҲҳ', 'мҳҒнқ¬', 'м§ҖнҳҒ', 'нҳңмӣҗ'];

students.forEach(name => {
  console.log(`${name}лӢҳ нҷҳмҳҒн•©лӢҲлӢӨ!`);
});
```

## мҳҲм ң 2-2: map - мғҒн’Ҳ к°ҖкІ© 10% мқёмғҒн•ҳкё°

**лҚ°мқҙн„° мӨҖл№„:**

products ліҖмҲҳм—җ мғҒн’Ҳ мқҙлҰ„кіј к°ҖкІ©мқҙ м ҖмһҘлҗң л°°м—ҙ к°қмІҙмқҳ мЈјмҶҢлҘј н• лӢ№н•ңлӢӨ.

**лҚ°мқҙн„° мІҳлҰ¬:**

1. map н•ЁмҲҳлҘј мӮ¬мҡ©н•ҳм—¬ products л°°м—ҙмқ„ мҲңнҡҢн•ҳл©° мғҒн’Ҳ к°ҖкІ© мҡ”мҶҢм—җ 10% мқёмғҒлҗң к°’мқ„ мһ¬н• лӢ№н•ңлӢӨ.

2. мң„м—җм„ң мІҳлҰ¬н•ң кІ°кіјлҘј мғҲлЎңмҡҙ л°°м—ҙлЎң л°ҳнҷҳн•ҳм—¬ raisedProducts ліҖмҲҳм—җ м ҖмһҘн•ңлӢӨ.

3. raisedProductsмҷҖ products л°°м—ҙмқ„ мҪҳмҶ”м°Ҫм—җ м¶ңл Ҙн•ҳм—¬ ліҖкІҪ мӮ¬н•ӯмқ„ л№„көҗн•ҳкі  нҷ•мқён•ңлӢӨ.

```python
%%writefile ex02_2.js

const products = [
  { name: 'м»Өн”ј', price: 4000 },
  { name: 'лқјл–ј', price: 4500 },
  { name: 'мјҖмқҙнҒ¬', price: 6000 }
];

const raisedProducts = products.map(item => ({
  ...item,
  price: item.price * 1.1
}));

console.log(raisedProducts);
console.log(products[0].price);
```

## мҳҲм ң 2-3: filter - мһ¬кі к°Җ мһҲлҠ” мғҒн’Ҳл§Ң кіЁлқјлӮҙкё°

**лҚ°мқҙн„° мӨҖл№„:**

inventory ліҖмҲҳм—җ мғҒн’Ҳ id, мқҙлҰ„, мһ¬кі  нҳ„нҷ© лҚ°мқҙн„°к°Җ м ҖмһҘлҗң л°°м—ҙ к°қмІҙ мЈјмҶҢлҘј н• лӢ№

**лҚ°мқҙн„° мІҳлҰ¬:**

1. filter н•ЁмҲҳлҘј мӮ¬мҡ©н•ҙ inStock мҡ”мҶҢмқҳ к°’мқҙ trueмқё н–үл§Ң н•„н„°л§Ғн•ҳм—¬ мғҲлЎңмҡҙ л°°м—ҙ к°қмІҙ л°ҳнҷҳ

2. л°ҳнҷҳлҗң л°°м—ҙмқҳ мЈјмҶҢлҘј availableItems ліҖмҲҳм—җ м ҖмһҘн•ҳм—¬ мҪҳмҶ”м°Ҫм—җ м¶ңл Ҙ

```python
%%writefile ex02_3.js

const inventory = [
  { id: 1, name: 'м•„мқҙнҸ°', inStock: true },
  { id: 2, name: 'л§Ҙл¶Ғ', inStock: false },
  { id: 3, name: 'м•„мқҙнҢЁл“ң', inStock: true },
  { id: 4, name: 'м—җм–ҙнҢҹ', inStock: false }
];

const availableItems = inventory.filter(item => item.inStock);

console.log(availableItems);
```

```python
!node ex02_3.js
```

## мҳҲм ң 2-4: reduce - мһҘл°”кө¬лӢҲ мҙқ кІ°м ң кёҲм•Ў кі„мӮ°

**лҚ°мқҙн„° мӨҖл№„:**

cart ліҖмҲҳм—җ мғҒн’ҲлӘ…/мғҒн’Ҳ к°ҖкІ© лҚ°мқҙн„° м ҖмһҘлҗң л°°м—ҙ к°қмІҙ мЈјмҶҢ н• лӢ№

**лҚ°мқҙн„° мІҳлҰ¬:**

1. reduce н•ЁмҲҳлҘј нҶөн•ҙ cart л°°м—ҙмқ„ мҲңнҡҢн•ҳл©° мғҒн’Ҳ к°ҖкІ© мҡ”мҶҢмқҳ мҙқн•©мқ„ кі„мӮ°н•ҳлҠ” лҸҷмһ‘мқ„ мҲҳн–ү

2. м—°мӮ° кІ°кіјлЎң м–»мқҖ к°’мқ„ totalPrice ліҖмҲҳм—җ н• лӢ№

3. totalPrice к°’мқ„ л©”мӢңм§ҖмҷҖ н•Ёк»ҳ мҪҳмҶ”м°Ҫм—җ м¶ңл Ҙ

```python
%%writefile ex02_4.js

const cart = [
  { name: 'м…”мё ', price: 35000 },
  { name: 'л°”м§Җ', price: 48000 },
  { name: 'м–‘л§җ', price: 5000 }
];

const totalPrice = cart.reduce((acc, cur) => acc + cur.price, 0);

console.log(`мҙқ кІ°м ң кёҲм•Ў: ${totalPrice}мӣҗ`);
```

```python
!node ex02_4.js
```

## мҳҲм ң 2-5: мў…н•©(Chaining) - кі л“қм җмһҗ лӘ…лӢЁ к°Җкіөн•ҳкё°

**лҚ°мқҙн„° мӨҖл№„:**

scores ліҖмҲҳм—җ н•ҷмғқ мқҙлҰ„/м җмҲҳ лҚ°мқҙн„°к°Җ нҸ¬н•Ёлҗң л°°м—ҙ к°қмІҙмқҳ мЈјмҶҢлҘј н• лӢ№

**лҚ°мқҙн„° мІҳлҰ¬:**

1. filter н•ЁмҲҳлҘј нҶөн•ҙ scores л°°м—ҙмқ„ мҲңнҡҢн•ҳл©° м җмҲҳк°Җ 80м җ мқҙмғҒмқё н•ҷмғқ лҚ°мқҙн„°л§Ңмқ„ м¶”м¶ңн•ҳм—¬ мғҲлЎңмҡҙ л°°м—ҙлЎң л°ҳнҷҳ

2. мң„м—җм„ң л§Ңл“Өм–ҙм§„ мғҲлЎңмҡҙ л°°м—ҙм—җ м ‘к·јн•ҳм—¬ мқҙлҰ„ мҡ”мҶҢл§Ң м¶”м¶ң нӣ„ мғҲлЎңмҡҙ л°°м—ҙлЎң лӢӨмӢң л°ҳнҷҳ

3. мөңмў… мғқм„ұлҗң л°°м—ҙмқ„ к°ҖлӮҳлӢӨмҲңмңјлЎң м •л ¬н•ҳм—¬ honorsList ліҖмҲҳм—җ м ҖмһҘ

4. honorsList л°°м—ҙмқ„ мҪҳмҶ”м°Ҫм—җ м¶ңл Ҙ

```python
%%writefile ex02_5.js

const scores = [
  { name: 'к№ҖмІ мҲҳ', score: 70 },
  { name: 'мқҙмҳҒнқ¬', score: 95 },
  { name: 'л°•м§Җм„ұ', score: 88 },
  { name: 'мөңмң лҰ¬', score: 60 },
  { name: 'м •мҡ°м„ұ', score: 92 }
];

const honorsList = scores
  .filter(s => s.score >= 80)
  .map(s => s.name)
  .sort();

console.log(honorsList);
```

```python
!node ex02_5.js
```
