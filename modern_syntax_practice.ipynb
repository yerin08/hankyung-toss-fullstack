# 현대적 객체 다루기 (Modern Syntax)

## 1. 비구조화 할당 (Destructuring)

객체나 배열에 저장된 데이터 중 필요한 값만 쏙 뽑아내어 변수에 할당하는 기술이다.

```python
%%writefile pt01.js

// 1. 객체 비구조화
const user = { id: 1, nickname: 'Gemini', email: 'ai@google.com' };
const { nickname, email } = user;
console.log(nickname); // 'Gemini' 출력

// 2. 배열 비구조화
const colors = ['red', 'green', 'blue'];
const [primary, secondary] = colors;
console.log(primary); // 'red' 출력
```

```python
!node pt01.js
```

### 실습 1. 객체 비구조화 할당 - 프로필 정보 뽑기

사용자 객체에서 필요한 정보만 변수로 추출하여 인사말을 완성하세요.

```python
%%writefile ex01.js

const user = {
  id: 1,
  displayName: '별명왕',
  email: 'king@example.com',
  level: 5
};

// 비구조화 할당을 사용하여 displayName과 level을 추출
const { displayName, level } = user;

console.log(`${displayName}님의 현재 레벨은 ${level}입니다.`);
```

```python
!node ex01.js
```

### 실습 2. 배열 비구조화 할당 - 순위 데이터 다루기

경주 결과 배열에서 1등과 2등을 변수에 담고, 나머지는 '나머지 선수들' 배열로 묶으세요.

```python
%%writefile ex02.js

const racers = ['김철수', '이영희', '박지성', '최유리', '정우성'];

// 비구조화 할당으로 1등, 2등, 그리고 나머지를 분리
const [gold, silver, ...others] = racers;

console.log(`금메달: ${gold}, 은메달: ${silver}`);
console.log(`나머지 선수들: ${others.join(', ')}`);
```

```python
!node ex02.js
```

## 2. 단축 속성명과 템플릿 리터럴

### 2-1. 단축 속성명 (Property Shorthand)

객체를 만들 때 변수 이름과 객체의 키 이름이 같다면 한 번만 써도 된다.

```python
%%writefile pt02_1.js

const name = 'Steve';
const age = 25;

// 과거: { name: name, age: age }
const person = { name, age };
console.log(person);
```

```python
!node pt02_1.js
```

### 2-2. 템플릿 리터럴 (Template Literals)

따옴표 대신 백틱(`)을 사용하여 문자열 속에 변수를 직접 삽입한다.

```python
%%writefile pt02_2.js

const product = 'MacBook';
const price = 2000000;

// 플러스(+) 기호 없이 직관적으로 작성
console.log(`${product}의 가격은 ${price.toLocaleString()}원 입니다.`);
```

```python
!node pt02_2.js
```

### 실습 3. 단축 속성명 & 템플릿 리터럴 - 동적 메시지 생성

변수들을 객체로 묶고, 그 데이터를 사용하여 한 줄의 문장을 만드세요.

```python
%%writefile ex03.js

const category = '가전';
const productName = '건조기';
const price = 1500000;

// 단축 속성명을 사용하여 item 객체 만들기
const item = { category, productName, price };

// 템플릿 리터럴을 사용하여 콘솔창에 정보 출력
console.log(`[${item.category}] ${item.productName}의 가격은 ${item.price.toLocaleString()}원 입니다.`);
```

```python
!node ex03.js
```

## 3. 안전한 코드 작성을 위한 방어 문법



### 3-1. 선택적 체이닝 (Optional Chaining, ?.)

속성이 존재하는지 확인하지 않고도 깊숙한 곳의 값에 접근할 수 있다. 만약 앞의 값이 null이나 undefined라면 에러를 내는 대신 undefined를 반환하고 멈춘다.

```python
%%writefile pt03_1.js

const order = { id: 101, info: {date: '2024-05-20'} };

// info가 없을 수도 있는 상황이라면?
console.log(order.info?.date); // '2024-05-20' 출력
console.log(order.user?.name); // 에러 없이 undefined 출력
```

```python
!node pt03_1.js
```

### 3-2. Null 병합 연산자 (Nullish Coalescing, ??)

값이 null이거나 undefined일 때만 사용할 **기본값(Default value)**을 지정한다.

*  주의: ||(OR 연산자)와 달리 0이나 ""(빈 문자열)은 유효한 값으로 취급하여 기본값으로 덮어쓰지 않는다.

```python
%%writefile pt03_2.js

let userNickname = null;
let displayName = userNickname ?? '익명 사용자'; // '익명 사용자'

let score = 0;
let finalScore = score ?? 100; // 0 (0은 유효한 값으로 인정)

console.log(displayName);
console.log(finalScore);
```

```python
!node pt03_2.js
```

### 실습 4. 선택적 체이닝(?.) - 중첩 객체 안전하게 접근하기

서버에서 받아온 데이터가 불완전할 때, 에러 없이 도시(city) 이름을 출력하세요.

```python
%%writefile ex04.js

const response = {
  user: {
    name: '홍길동',
    // address 정보가 없을 수도 있음
  }
};

// 선택적 체이닝을 사용하여 address 안의 city에 접근
const city = response.user.address?.city;

console.log(city); // 에러 없이 undefined 출력
```

```python
!node ex04.js
```

### 실습 5. 종합 문제 - 사용자 설정 기본값 적용

사용자가 설정한 테마나 폰트 크기가 없을 때만 기본값을 적용하는 로직을 작성하세요.


* 조건: ?? (Null 병합 연산자)를 사용하여 0이나 false 값은 유효하게 처리하세요.



```python
%%writefile ex05.js

const userSettings = {
  theme: null,
  fontSize: 0, // 사용자가 가장 작게 설정한 상태
  isAlarmOn: false
};

// Null 병합 연산자를 사용하여 값이 없을 때만 기본값 할당
const theme = userSettings.theme ?? 'light';
const fontSize = userSettings.fontSize ?? 16; // 0으로 값이 설정되어 있기 때문에 0이 유지되어야 함
const isAlarmOn = userSettings.isAlarmOn ?? true; // 위와 같은 이유로 초기 설정값 false가 유지되어야 함

console.log(`테마: ${theme}, 폰트: ${fontSize}, 알람: ${isAlarmOn}`);
```

```python
!node ex05.js
```
